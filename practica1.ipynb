### Operación entre Vectores
Importar la libreria 'numpy' para realizar operaciones matemáticas con vectores y matrices. Existen dos maneras de realizar la multiplicación escalar, con la función 'np.dot' y el operador '@'. La multiplicación punto a punto se realiza con el operador '*'.

```python
import numpy as np

# Creacion de vectores
a=np.array([3.1,1,-0.5,-3.2,6])
b=np.array([1,3,2.2,5.1,1])

# Multiplicacion escalar 
m_escalar = np.dot(a,b)
print("La multiplicación escalar de a y b es:", m_escalar)

# Multiplicacion punto a punto
m_punto_punto = a*b
print("La multiplicación punto a punto de a y b es:", m_punto_punto)
```

### Operacion entre matrices
Funciones:
* Ones: Crea un array con las dimensiones especificadas en la tupla dimensiones cuyos elementos son todos unos.
* Round: Redondea el valor decimal de un número a su entero más cercano.
* Ceil: Devuelve el número entero más próximo al pasado como argumento que sea mayor o igual a él.
* Floor: Devuelve el número entero más próximo al pasado como argumento que sea menor o igual a él.

### Operacion entre funciones
La función 'np.arange' crea un array de valores igualmente espaciados en un rango especifico. La biblioteca 'matplotlib' se utiliza para construir y graficar señales 

```python
import matplotlib.pyplot as plt

# Construccion del intervalo
n = np.arange(0, 101)  # n desde 0 hasta 100

# Construcción de las señales seno con diferentes frecuencias
y1 = np.sin(np.pi * 0.12 * n)
y2 = np.sin(2 * np.pi * 0.03 * n)

# Tercera señal. Suma de y1 y y2
s = y1 + y2

# Cuarta señal. Poducto de y1 y y2
t = y1 * y2

# Graficar y1[n] y y2[n]
plt.figure()
plt.plot(n, y1, label='y1[n]', color='b')
plt.plot(n, y2, label='y2[n]', color='r')
plt.title('Señales y1[n] y y2[n]')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)

# Graficar s[n] y t[n]
plt.figure()
plt.plot(n, s, label='s[n] (Suma)', color='g')
plt.plot(n, t, label='t[n] (Producto)', color='m')
plt.title('Señales s[n] y t[n]')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)

plt.show()
```

### RETO
Escribir una función que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con la nota mínima, la máxima, media y la desviación típica.

```python
%reset
import statistics as stc
import pandas 
from pandas import Series

alumnos = {
    'Laura': [1,2,3,4,5,6],
    'Maria': [7,8,9,2,3,4],
    'Otis': [1,3,5,7,9,1],
    'Chloe': [2,4,6,8,4,6]
}

while True:
    name = input("Ingrese el nombre del alumno: ")
    if name in alumnos:
        buscar = alumnos[name]
        print(buscar)
        maximo = max(buscar)
        minimo = min(buscar)
        media = stc.mean(buscar)
        des = stc.stdev(buscar)
        serie = Series([minimo,maximo,media,des], index= ["Nota minima","Nota maxima","Media","Desviacion tipica"])
        print(serie)
        break
    else :
        print("No existe ese alumno")
```
