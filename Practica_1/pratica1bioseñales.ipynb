```python
import numpy as np
import matplotlib.pyplot as plt

# a vectores
a = np.array([3.1, 1, -0.5, -3.2, 6])
b = np.array([1, 3, 2.2, 5.1, 1])

#multiplicación escalar (producto punto) de a y b
dot_product = np.dot(a, b)
print(f'Multiplicación escalar (a.b): {dot_product}')

#multiplicación punto a punto de a y b
pointwise_product = a * b
print(f'Multiplicación punto a punto (a * b): {pointwise_product}')

#matriz A
A = np.array([[2, -1, -3],
              [4, 1.5, -2.5],
              [7.3, -0.9, 0.2]])

#la transpuesta de A
A_transpose = np.transpose(A)
print(f'Transpuesta de A:\n{A_transpose}')

#funciones de numpy
print(f'np.ones(3): {np.ones(3)}')
print(f'np.round(3.456, 1): {np.round(3.456, 1)}')
print(f'np.ceil(3.2): {np.ceil(3.2)}')
print(f'np.floor(3.8): {np.floor(3.8)}')

#valor de la primera fila, tercera columna de la matriz A
valor = A[0, 2]
print(f'Valor de la primera fila, tercera columna: {valor}')

#segunda fila de la matriz A
segunda_fila = A[1, :]
print(f'Segunda fila de la matriz A: {segunda_fila}')

#dimensiones de la matriz A
dimensiones = A.shape
print(f'Dimensiones de la matriz A: {dimensiones}')

#función yn = sin(0.12 * n) en el intervalo 0 <= n <= 100
n = np.arange(0, 101)
yn = np.sin(0.12 * n)

#función y2n = cos(2 * 0.03 * n)
y2n = np.cos(2 * 0.03 * n)

#tercera señal sn = yn + y2[n]
sn = yn + y2n

#cuarta señal tn = yn * y2[n]
tn = yn * y2n

#graficar y[n] y y2[n]
plt.figure(figsize=(10, 5))
plt.subplot(2, 1, 1)
plt.plot(n, yn, label='y[n] = sin(0.12n)', color='blue')
plt.plot(n, y2n, label='y2[n] = cos(2*0.03n)', color='orange')
plt.legend()
plt.title('Señales y[n] y y2[n]')
plt.xlabel('n')
plt.ylabel('Amplitud')

#graficar s[n] y t[n]
plt.subplot(2, 1, 2)
plt.plot(n, sn, label='s[n] = y[n] + y2[n]', color='green')
plt.plot(n, tn, label='t[n] = y[n] * y2[n]', color='red')
plt.legend()
plt.title('Señales s[n] y t[n]')
plt.xlabel('n')
plt.ylabel('Amplitud')

plt.tight_layout()
plt.show()


```

```python
import pandas as pd
import numpy as np

def calcular_estadisticas_notas(notas):

    # Convertir el diccionario a una serie de pandas
    notas_serie = pd.Series(notas)

    # Calcular las estadísticas
    min_nota = notas_serie.min()
    max_nota = notas_serie.max()
    media_nota = notas_serie.mean()
    desviacion_tipica = notas_serie.std()

    # Verificar cuántos alumnos aprobaron
    aprobados = (notas_serie >= 3.0).sum()
    total_alumnos = len(notas_serie)
    porcentaje_aprobados = (aprobados / total_alumnos) * 100

    # Crear una serie con las estadísticas
    estadisticas = pd.Series({
        'Mínima': min_nota,
        'Máxima': max_nota,
        'Media': media_nota,
        'Desviación Típica': desviacion_tipica,
        'Aprobados (%)': porcentaje_aprobados
    })

    return estadisticas

notas_alumnos = {
    'Juan': 4.5,
    'Maria': 3.2,
    'Juliana': 2.9,
    'Cristian': 4.8,
    'Luis': 3.0
}

resultados = calcular_estadisticas_notas(notas_alumnos)
print(resultados)
```
