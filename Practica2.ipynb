# Impulso Unitario

```python
def impseq(n0,n1,n2):
  n=np.arange(n1,n2+1)
  x=(n-n0)==0
  return [x,n]
```

```python
plt.stem(impseq(2,1,8)[0])
```

# Escalón unitario

```python
def unit(n0,n1,n2):
  n=np.arange(n1,n2+1)
  x=(n-n0)>=0
  return [x,n]
```

```python
plt.stem(unit(2,1,8)[0])
```

# Rampa

```python
def rampa(n0,n1,n2,m):
  n=np.arange(n1,n2+1)
  x1=(n-n0)>0
  x2=np.array(m*(n-n0))
  x=x1*x2
  return [x,n]
```

```python
plt.stem(rampa(2,1,8,1)[0])
```

# Exponenciales complejas


```python
def F_exp(n,At,w0):
    x_e=np.exp((At+1j*w0)*n)
    return x_e
```

```python
plt.stem(F_exp(n,-0.5,5))
```

# Sinusoides

```python
def F_sin(n,w0,theta):
    x_s=np.sin(n*w0+theta)
    return x_s
```

```python
plt.stem(F_sin(n,5,15))
```

## Genere la siguiente secuencia


```python
rn=rampa(0,0,5,1)[0]
rn5=rampa(5,6,11,1)[0]
rn10=rampa(10,12,17,1)[0]
xn=np.hstack((rn,rn5,rn10))
n=np.linspace(0,np.shape(xn)[0]-1,np.shape(xn)[0])
plt.stem(n,xn)
```

### a.	Derive la secuencia 

```python
dxn = np.zeros_like(xn)
dxn[1:6] = xn[1:6] - xn[0:5]  
dxn[7:12] = xn[7:12] - xn[6:11] 
dxn[13:18] = xn[13:18] - xn[12:17] 
```

### B. Subplot con las dos secuencias

```python
plt.figure(figsize=(10, 6))

# Graficar x[n]
plt.subplot(2, 1, 1)
plt.stem(n, xn)
plt.title('Secuencia Original x[n]')
plt.xlabel('n')
plt.ylabel('x[n]')

# Graficar la derivada de x[n]
plt.subplot(2, 1, 2)
plt.stem(n, dxn)
plt.title('Derivada de x[n]')
plt.xlabel('n')
plt.ylabel('dx[n]/dn')

plt.tight_layout()
plt.show()
```
